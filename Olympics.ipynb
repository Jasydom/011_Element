**Step I : Import libriaries and dataset**

```python
import pandas as pd
import numpy as np
import tensorflow 
import matplotlib.pyplot as plt
import scipy as sc
import seaborn as sns
dataset = pd.read_csv("olympics-3.csv")
```

**Step II : Data Encoding**

```python
dataset["Medal"] = dataset["Medal"].replace('Gold', 1)
dataset["Medal"] = dataset["Medal"].replace('Silver', 2)
dataset["Medal"] = dataset["Medal"].replace('Bronze', 3)
```

```python
dataset["Event_gender"] = dataset["Event_gender"].replace('M', 0)
dataset["Event_gender"] = dataset["Event_gender"].replace('W', 1)
dataset["Event_gender"] = dataset["Event_gender"].replace('X', 2)
```

```python
dataset["Gender"] = dataset["Gender"].replace('Men', 1)
dataset["Gender"] = dataset["Gender"].replace('Women', 0)
```

**Step III: Pie chart on the distribution of men and women for golden medal**

```python
result = dataset.groupby('Gender')['Athlete'].nunique()
f = plt.figure(figsize=(7, 7))
colors = ["red","blue"]
plt.pie(result, colors = colors, autopct = '%0.0f%%')
plt.title('Pie on the distribution of men and women', fontsize=20)
plt.legend(['Women','Men'],fontsize = 12)
plt.show(f.number)
```

Men won three times more medals than women.

**Step IV: Pie chart on the distribution of different type event for gender**

```python
result = dataset.groupby('Event_gender')['Athlete'].nunique()
f = plt.figure(figsize=(7, 7))
colors = ["red","blue","orange"]
plt.pie(result, colors = colors, autopct = '%0.0f%%')
plt.title('Pie on the distribution of men and women', fontsize=20)
plt.legend(['Men','Women','Mixed'],fontsize = 12)
plt.show(f.number)
```

**Step V : Histogram on the number of golden medal received by country (Top 5)**

```python
National_Olympic_Committee = dataset['NOC'].value_counts().nlargest(5)
largeur_barre = 0.8 
names = ['USA', 'URS', 'GBR', 'FRA', 'ITA']
x1 = range(len(names)) 
f = plt.figure(figsize=(12, 10))
plt.title("Number of medal",fontsize = 18,fontweight='bold')
plt.bar(x1,National_Olympic_Committee,width=0.8, color = '#2FBDE7', linewidth = 2)
plt.xticks([r + largeur_barre / 2 for r in range(len(x1))],names,fontsize = 13,rotation=45)
plt.legend(['Number of medal'],fontsize = 12)
plt.show()
```
